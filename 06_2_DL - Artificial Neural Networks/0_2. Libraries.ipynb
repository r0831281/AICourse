## 1. Tensorflow

TensorFlow is a powerful data flow oriented machine learning library created by the Brain Team of Google and made Open Source in 2015. It was based on the Theano library, and is designed to be easy to use and widely applicable to both numeric and neural network oriented problems as well as other domains.

## 2. Keras

Keras is an open-source library for building Neural Networks at a high level in Python, i.e. Keras is the high-level API of TensorFlow. In practical terms, Keras makes implementing the many powerful but often complex functions of TensorFlow as simple as possible, and it is configured to work with Python without any major modifications or configuration. It is minimalistic and straightforward with a high level of extensibility. The minimalistic approach in design aims at fast and easy experimentation through the building of compact systems. Keras is really eased to get started with, and it will be our choice for this part of the course.

<img src="./resources/keras.png"  style="height: 200px"/>

## 3. PyTorch

PyTorch is a framework developed by Facebook in 2017, based on the Torch library. It uses a dynamic computational graph, which means that the graph is built on the fly as the code is executed (as compared to Tensorflow, where we first have to compile our model, before we send data through it). 

PyTorch will be the platform for a follow up course on Deep Learning. In this course we'll be using Keras, because of it's great documentation, and readable code.
